```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
social_media_hours = np.random.randint(1, 5, 100)
social_media_hours
```

```python
gaming_hours = np.random.randint(1, 5, 100)
gaming_hours
```

```python
battery_drain = social_media_hours * 3.5 + gaming_hours * 5.7 + np.random.normal(0, 0.5, 100)
battery_drain
```

```python
# battery drain to 2 decimal points
battery_drain = np.round(battery_drain, 2)
battery_drain
```

```python
data = pd.DataFrame({
    'social_media_hours': social_media_hours,
    'gaming_hours': gaming_hours,
    'battery_drain': battery_drain
})
data.head()
```

```python
x = data[['social_media_hours', 'gaming_hours']]
y = data['battery_drain']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
plt.scatter(x_train, y_train, label = "Test Data")
# plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', label = "Predicted Data")
plt.plot(x_train, model.predict(x_train.values.reshape(-1, 1)), color='red', label='Regression Line')
plt.xlabel('Actual Battery Drain')
plt.ylabel('Predicted Battery Drain')
plt.title('Actual vs Predicted Battery Drain')
plt.show()
```
